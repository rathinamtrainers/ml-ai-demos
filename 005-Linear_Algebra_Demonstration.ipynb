```python
import numpy as np
```

```python
t = np.linspace(0,40,1000)
```

```python
print(t)
```

```python
d_robber = 2.5 * t
```

```python
d_police = 3 * (t - 5)
```

```python
d_robber
```

```python
d_police
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
plt.title("Bank Robber Chasing")
plt.xlabel('time (in minutes)')
plt.ylabel("distance (in km)")
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_robber, c='green')
ax.plot(t, d_police, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple', linestyle='--')
```
